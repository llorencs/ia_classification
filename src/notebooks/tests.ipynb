```python
import pandas as pd
```

```python
file = r'C:\Projects\Python\AI\classification\files\Language Detection.csv'
df = pd.read_csv(file)
```

```python
df.head(10)
```

```python
df['Text']
```

```python
from sklearn import svm
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(df['Text'])
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['Language'])
```

```python
clf = svm.SVC()
clf.fit(x,y)
```

```python
data = {
    'Text': [
        "This is a test sentence in English.",
        "Esta es una oración de prueba en español.",
        "Machine learning is fascinating.",
        "El aprendizaje automático es fascinante.",
        "Hello, world!",
        "¡Hola, mundo!",
        "Data science involves statistics, mathematics, and computer science.",
        "La ciencia de datos involucra estadísticas, matemáticas y ciencias de la computación.",
        "Artificial intelligence will change the world.",
        "Искусственный интеллект изменит мир."
    ],
    'Language': [
        "English",
        "Spanish",
        "English",
        "Spanish",
        "English",
        "Spanish",
        "English",
        "Spanish",
        "English",
        "Russian"
    ]
}

# Creating the DataFrame
test_df = pd.DataFrame(data)
```

```python
x_test = vectorizer.transform(test_df['Text'])
predictions = clf.predict(x_test)
```

```python
predicted_labels = label_encoder.inverse_transform(predictions)
```

```python
print(predicted_labels)
```

```python
y_test = label_encoder.transform(test_df['Language'])
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_pred, labels=range(len(label_encoder.classes_)), target_names=label_encoder.classes_))
```
